```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Dense, Embedding, LSTM, SpatialDropout1D
from sklearn.model_selection import train_test_split
from keras.utils.np_utils import to_categorical
from keras.callbacks import EarlyStopping
from keras.layers import Dropout
import re
```

```python
df = pd.read_csv('../dataset/Gungor_2018_VictorianAuthorAttribution_data-train.csv', encoding = "cp1252")
```

```python
max_words = 10000
# Max number of words in each complaint.
max_length = 1000
# This is fixed.
embedding_dim = 100
tokenizer = Tokenizer(num_words=max_words, filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~', lower=True)
tokenizer.fit_on_texts(df.text)
print("tokens:", tokenizer.word_index)
```

```python
train = tokenizer.texts_to_sequences(df.text)
train_data = pad_sequences(train)
```

```python
train_label = to_categorical(df.author - 1)
```

```python
train_data.shape
```

```python
train
```

```python
model = Sequential()
model.add(Embedding(max_words, embedding_dim, input_length=train_data.shape[1]))
model.add(SpatialDropout1D(0.2))
model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(50, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
epochs = 3
batch_size = 64
model.fit(train_data[:100], train_label[:100], epochs=epochs, batch_size=batch_size,validation_split=0.1)
```
