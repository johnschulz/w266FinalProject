```python
train_data = pd.read_csv('dataset/Gungor_2018_VictorianAuthorAttribution_data-train.csv', encoding='latin-1')
train_data 
```

Clean Data: Lemmatisation and Remove Stopwords

```python
lemmatiser = WordNetLemmatizer()
stop_words = stopwords.words('english')
cleaned_text = []
for text in train_data['text']:
    lemmatised_words =''
    split_text = text.split()
    for word in split_text:
        if word not in stop_words:
            lemmatised_word=lemmatiser.lemmatize(word, pos="v")
            lemmatised_words += lemmatised_word + ' '
    cleaned_text.append(lemmatised_words)
train_data['text'] = cleaned_text
train_data
```
